**DATA UNDERSTANDING**
This  involves data cleaning



```
#this involves changing the columns to UPPER case
df.columns=df.columns.str.upper()
df.columns
df.head(1)
```

```
#this involves changing the columns to UPPER case
df1.columns=df.columns.str.upper()
df1.columns
df.head(3)
```

```
#this involves changing the columns to lower case
df2.columns=df.columns.str.upper()
df2.columns
df1.head(1)
```

**COMBINING THE TWO DATASETS**

```
# this is to join the three datasets
df_combine=pd.concat([df,df1,df2])
df.combine
df.tail(3)
```

```
#drop the columns that we dont need
df_cleaned= df_combine.drop(columns = ['COUNTRY_A',	'COUNTRY_B' ])
df_cleaned.head()
```

```
# we then drop duplicates in our combined dataset
df_cleaned1=df_cleaned.drop_duplicates()
df_cleaned1.tail(2)
```

```
#after dropping the duplicates we preview our table
df_cleaned1.info()
```

```

# merging the telecom datasets and preveiwing it
df_cleaned1=pd.merge(df_cleaned1,df3,how='inner', left_on='SITE_ID',right_on='SITE_CODE')
df_cleaned1.head(3)
```

```
#split the date and time columns for ease of calculation
df_cleaned1[['DATE', 'TIME']] = df_cleaned1.DATETIME.str.split(expand=True)
df_cleaned1.head(3)
```

**DATA ANALYSIS**
We use our data to obatin insights and make conclusion 


```
#to find the most popular product in mtn
df_cleaned1.groupby('PRODUTC')['VALUE'].sum()
```

```
#finding the most popular cell id in the region
df_cleaned1.groupby('CELL_ID').count().nlargest(3,'PRODUTC')
```

```
#the most popular site id in the region
df_cleaned1.groupby('SITE_ID').count().nlargest(3,'PRODUTC')
```

```

#find the busiest time in ivoty coast
pd.set_option("display.max_rows",None)
TIME = df_cleaned1.groupby(df_cleaned1.index.TIME).mean()
TIME
TIME.rank(method='max')
df_cleaned1.resample("H").sum()
```
